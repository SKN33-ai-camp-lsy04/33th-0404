이를 통해 코드 처리흐름을 명확히 전달할 수 있다.

```python
def foo():
    raise NotImplementedError("foo() 함수가 구현되지 않았습니다")


class AdultGame:
    def __init__(self):
        pass
    def start(self):
        print("성인용 게임(고스톱)을 시작합니다.")

# if - else

age = int(input("나이 입력: "))

if age >= 19:
        AdultGame().start()
else:
        print("go away")


# try-except

```

```python
class UnderAgeError(Exception):
    pass
# 기존 exceptiion,Error 클래스 중 아무거나 상속 받기

class AdultGame2:
    def __init__(self, age: int):
        if age <19 :
            # Baseexception :모든 예외, 에러 클래스의 최상위 부모 클래스
            # raise BaseException(f"미성년자는 이용할 수 없습니다: {age}")

            # 상황에 맞는 예외 강제 발생 (권장)
            # 파이썬이 제공해주는 알맞은 예외가 없으면 사용자가 생성 가능:
            raise UnderAgeError (f"미성년자는 이용할 수 없습니다: {age}")

    def start(self):
        print("성인용 게임(고스톱)을 시작합니다.")


try:
    age = int(input("나이 입력: "))
    AdultGame2(age).start()
except UnderAgeError as e:
    print(type(e).__name__)
    print(e)

print("종료")
```
