```python
# except 구문을 이용한 예외 객체 사용 방법
import random

try:
    num = random.randint(0,1)   # 0~1 사이 정수형 난수 생성
    print("num: ", num)

# 0 == False, 1== True
    if num: # 1인 경우
        print('abc' + 123)
    else:   # 0인 경우
        print(int('abc'))
except TypeError as e:  # 별칭 만들기
    # e: try에서 발생한 예외 객체
    print(e)
    print("예외 타입 이름: ", type(e).__name__)
    print("예외 메시지: ", str(e))
    print("예외 상세값: ", e.args)
    print(f'[Error] {type(e).__name__}: {str(e)}')

except ValueError as e:
    print(e)
    print("예외 타입 이름: ", type(e).__name__)
    print("예외 메시지: ", str(e))
    print("예외 상세값: ", e.args)
    print(f'[Error] {type(e).__name__}: {str(e)}')
```

## 예외처리
**(중요) 발생된 예외/오류 상황에서 try..except구문을 통해 적절히 처리후 정상적인 흐름을 돌려내는 것**

### pass
- if, for, def, class, except 등의 블록은 내부가 비어있으면 오류가 발생함
- 이 때, 구조만 유지하는 용도 또는 예외를 무시할 때 사용

```python
# 숫자를 입력받아
# 입력받은 숫자와 key가 일치하는 학생을 조회하여 id, name 출력
# 입력받은 숫자가 students에 없으면 "1~3 사이 숫자만 입력해주세요"
# + 잘못된 숫자 입력 시 다시 숫자를 입력 받게 하기

students: dict[int, dict] = {
    1: {"id": 10, "name": "홍길동"},
    2: {"id": 20, "name": "신사임당"},
    3: {"id": 30, "name": "이순신"},
}

while True:     #while 조건식 -> 조건식이 true인 경우 계속 반복
    try:
        num : int = int(input ("숫자입력: "))
        student: dict = students[num]
        print("id: ", student['id'])
        print("name: ",student['name'] )

        break

    except KeyError as e:
        print("1~3 사이 숫자만 입력해주세요")


print("--- 프로그램 종료 ---")
```

```python
# 예외 무시
try:
    print(10/0)
except:
    pass

print("종료")
```

```python
###  finally
#- try 구문에서 예외 발생 여부와 관게 없이
#   마지막에 무조건 실행하는 구문(블록)

### else
# except 구문을 모두 지나친 경우 수행되는 구문
```

```python
try:
    num = int(input("숫자 입력"))
    result = 100/num

except ValueError as e:
    print("정수형 숫자만 입력하세요")

except ZeroDivisionError as e:
    print("0으로 나눌 수 없습니다")

else:
    print("계산결과: ", result)

finally:
    print("예외 발생 여부와 관계 없이 무조건 수행")
```

```python
# finally 사용 예시

try:
    f = open('helloworld.txt', 'r', encoding = 'utf-8')
    print(f.read())
    print(10/0)
except:
    pass
finally:
    f.close()    # 메모리 반환(필수)

```
