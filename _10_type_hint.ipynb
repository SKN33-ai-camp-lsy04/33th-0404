# Type Hint

- Python은 **동적 타이핑(dynamic typing)** 언어이다.
- 변수 이름 자체에 자료형이 고정되는 것이 아니라, 변수가 가리키는 객체가 타입을 가진다.
- **type hint**는 코드 작성 시 IDE와 정적 분석 도구가 타입을 추론하고 경고할 수 있도록 돕는 표시이다.
- 중요한 점은 type hint가 Python 런타임에서 타입을 강제로 막아 주지는 않는다는 것이다.

## 1. 동적 타이핑

Python에서는 같은 변수 이름에 서로 다른 타입의 객체를 다시 대입할 수 있다.

- `x = '안녕'`일 때 `x`는 문자열 객체를 가리킨다.
- `x = 123`으로 다시 대입하면 `x`는 정수 객체를 가리킨다.
- 타입은 변수 이름이 아니라, 변수 이름이 가리키는 객체가 가진다.

## 4. 컬렉션 Type Hint

`list`, `tuple`, `dict`, `set` 같은 컬렉션에도 type hint를 작성할 수 있다.

- `list[int]`: 정수를 담는 리스트
- `tuple[int, str, str]`: 첫 번째는 정수, 두 번째와 세 번째는 문자열인 튜플
- `dict[str, str | int]`: key는 문자열, value는 문자열 또는 정수인 딕셔너리
- `set[str]`: 문자열을 담는 집합

````python
names: list[str] = ['홍길동', '이미아']
user: tuple[int,str, str] = (1, 'user01', 'user02')
info: dict[str, str | int | float] = {'num': 100, "s":'aaa'}
chars :set[str] = {'a','b','c'}
````

## 5. 함수 Type Hint

함수의 매개변수와 반환값에도 type hint를 작성할 수 있다.

```python
def 함수명(매개변수: 타입) -> 반환타입:
    ...
```

이 표기는 함수 사용법을 문서처럼 보여 주고, IDE나 정적 분석 도구가 잘못된 사용을 찾는 데 도움을 준다.

````python
def add(a: int, b: int): # 매개 변수에 힌트 작성 + 반환타입 힌트 작성
    return a+b

print(add(3,5))
print(add('3','5'))
````

## 6. 상수 표현과 Final

Python에는 값을 절대 바꾸지 못하게 만드는 진짜 상수 문법이 따로 있는 것은 아니다.

- 관례상 대문자로 작성된 변수는 상수처럼 다룬다.
- `typing.Final`은 이 값이 재대입되지 않아야 한다는 힌트를 제공한다.
- `Final`도 정적 분석 도구를 위한 힌트이며, Python 런타임이 재대입을 직접 막지는 않는다.
