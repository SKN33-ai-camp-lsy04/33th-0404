```python
# 출석부 예시 -- 출석 여부 바꿀 수 있게

test = [
    {'no':1, 'name':'홍길동', 'check':False},
    {'no':2, 'name':'김길동', 'check':True},
    {'no':3, 'name':'고길동', 'check':True}
]

def change_check(name, check):
    for std in test:
        if std['name'] == name:
            std['check'] = check

change_check('고길동', False)
change_check('홍길동', True)

print(test)
```
