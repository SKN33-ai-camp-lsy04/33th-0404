```python
# 리스트 정리
minus = ['완화', '인하', '억제', '폐지']
plus = ['지원', '인센티브', '수립','확대','활성화']
neutral = ['추진', '마련', '안정화', '형평성', '합리화', '이행', '개편']

# 변수 정의
promises =[
    "무주택 서민에게 주거안정과 내집마련의 희망을 제공하겠습니다.",
    "노후 공동주택 재개발 활성화 지원",
    "공공임대주택 확대를 통한 주거안정 추진",
    "임차료 안정화를 위한 인센티브 제도 도입",
    "전월세 소득공제 및 최저주거기준 미달 가구 지원 확대",
    "보유세 세부담 완화 방안 수립·이행",
    "1주택자 취득세 인하 추진"
]

# 함수 설정
def get_analyze(promise):

    print("공약을 분석해보겠습니다: ", promise)

    lst = promise.split( )

actions = []
keywords = []

# 조건별 출력값 설정
for word in lst:

    if word in plus:
        actions.append(("확대, word"))

    elif word in minus:
        actions.append(("축소, word"))

    elif word in neutral:
        actions.append(("중립, word"))
    else:
        keywords.append(word)

# 돌아라 돌아
for promise in promises:
    get_analyze(promise)
```
