```python
class MyClass:
    def __init__(self, value):
        self.__value = value

    # getter
    # 외부에서 private 속성 값을 얻어가게 하는 메서드
    def get_value(self):
        return self.__value

my_cls = MyClass(100)
print("my_cls.get_value(): ", my_cls.get_value())

    # setter
    # 외부에서 private 속성 값을 변경하는 메서드
    def set_value(self, value):
        self.__value = value

print("200 세팅")
my_cls.set_value(200)
print("my_cls.get_value():", my_cls.get_value())
```
