```python

class Account:
    def __init__(self,  username, balance):
        self.__username = username
        self.__balance = balance

    def info(self):
        print(f'{self.__username}님의 현재 잔액은 {self.__balance}원 입니다.')

    def deposit(self, money):
        if money <= 0:
            print("입금액이 유효하지 않습니다.")
        else:
            self.__balance += money
            print("입금 금액: ", money, "원")
            print("현재 잔액: ", self.__balance, '원')

    def withdraw(self, money):
        if money <= 0:
            print("출금액이 유효하지 않습니다.")
        elif money > self.__balance:
            print("잔액이 부족합니다.")
        else:
            self.__balance -= money
            print("출금 금액: ", money, "원")
            print("현재 잔액: ", self.__balance, "원")

account1 = Account("kim", 10000)

account1.__init__("kim", 10000)

account1.deposit(5000)
account1.withdraw(3000)
```

```python
class SKNStudent:
    # 클래스 속성: 해당 클래스로 만들어진 모든 인스턴스가 공유하는 값
    course_name = "Family AI camp"

    # 생성자
    def __init__(self, nth, name):
        self.nth = nth
        self.name = name

    # 인스턴스 메서드 : 첫 번째 매개변수로 self를 받는 메서드
    # - 해당 인스턴스의 속성을 다룰 수 있음
    def greeting(self):
        print(f'안녕하세요. {self.nth}기 {self.name}입니다.')

    # 클래스 메서드 : 첫 번째 매개변수로 cls를 받는 메서드
    # - cls ==해당클래스 == SKNStudent
    @classmethod #데코레이터
    def good_morning(cls):
        print(f'{cls.course_name} 여러분, 좋은 아침입니다!')

    # 정적 메서드 (static method)
    # - self, cls 같은 참조 변수를 받지 않음
    # - 인스턴스 상태, 클래스 상태를 직접 다루지 않는 독립적인 기능에 사용한다
    @staticmethod
    def course_info(campus):
        if campus == '서초':
            print("서초점 주소: 서울시 서초구 ...")
        elif campus == "동작":
            print("동작점 주소: 서울시 동작구...")
        else:
            print("잘못 입력")

# SKAStudent 객체(인스턴스 생성)
kim = SKNStudent(33, '김유진')
choi = SKNStudent(33, '최경돈')

print(kim.nth, kim.name, kim.course_name)
print(choi.nth, choi.name, choi.course_name)

# 인스턴스 속성 변경
kim.nth = 34
print("--- 인스턴스 속성 변경 후 ---")
print(kim.nth, kim.name, kim.course_name)
print(choi.nth, choi.name, choi.course_name)

# 클래스 속성 변경
SKNStudent.course_name = "AI Camp"
print("--- 클래스 속성 변경 후 ---")
print(kim.nth, kim.name, kim.course_name)
print(choi.nth, choi.name, choi.course_name)

kim.greeting()

kim.good_morning()
SKNStudent.good_morning()
# 클래스명. 클래스메서드() <권장>

SKNStudent.course_info("강남")
# 클래스명.정적메서드명() <권장>
```

```python
class Student:
    school_name = "서일 초등학교" #클래스 속성

    def __init__(self, name, grade):    #생성자를 통해 객체 생성(student 바탕으로 만든 name, grade 객체)
        self.name = name
        self.grade = grade

park = Student("박철수", 3)
kim = Student("김영희", 4)
# 새로 만든 객체에 변수 입력

print("park.school_name:  ", park.school_name)
print("kim.school_name: ", kim.school_name)
print("Student.school_name: ", Student.school_name)

# 인스턴스.클래스 속성 = "값"
# -> 해당 인스턴스에 "클래스속성명"과 같은 "인스턴스 속성"이 추가된다
park.school_name = "서이 초등학교"
print(park.__dict__)
print(kim.__dict__)
print()

print("park.school_name:  ", park.school_name)
print("kim.school_name: ", kim.school_name)
print("Student.school_name: ", Student.school_name)
```
