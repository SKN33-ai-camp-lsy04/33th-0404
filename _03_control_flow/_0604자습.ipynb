```python
# operation
## if - else 문법
Halloween = input("무엇을 주겠습니까: ")
print('treat' if Halloween == "candy" else 'trick')

# Halloween을 input으로 잘못 씀
```

```python
## 입력값이 공란인지 확인하기
source = input("소스 이름을 입력하세요")
print('포테이토칩 맛: ' + source if source else '이름을 입력해주세요')

# source를 예제 변수 그대로 옮겨적음
# 변수 혼동 주의
```

```python
# 조건문 if
## 시험 점수가 90점 이상일 때만 커피 쿠폰 지급
alcohol_by_volume = int(input("도수를 입력하세요: "))
print('drink' if alcohol_by_volume <= 16 else 'kids beverage')

# <= (순서 조심)
```

```python
## 점수별 합격 불합격

def get_season(degree):
    if degree >= 20:
        clothes = "반팔"
    elif degree >= 15:
        clothes = "긴팔"
    elif degree >= 10:
        clothes = "가디건"
    elif degree >= 0:
        clothes = "코트"
    else:
        clothes = "패딩"

    if not ( 0 <= degree <= 30):
        print("살아남아보세요")

    return clothes

degree = int(input("오늘의 최고 온도를 입력하세요: "))
clothes = get_season(degree)
print("오늘 입을 옷을 알려드립니다", clothes)
```
