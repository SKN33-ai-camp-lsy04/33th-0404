```python
## 점수별 합격 불합격

def get_season(degree):
    if degree >= 20:
        clothes = "반팔"
    elif degree >= 15:
        clothes = "긴팔"
    elif degree >= 10:
        clothes = "가디건"
    elif degree >= 0:
        clothes = "코트"
    else:
        clothes = "패딩"

    if not ( 0 <= degree <= 30):
        print("살아남아보세요")

    return clothes

degree = int(input("오늘의 최고 온도를 입력하세요: "))
clothes = get_season(degree)
print("오늘 입을 옷을 알려드립니다", clothes)
```

```python
# match_case
## 표현식 case별 실행문
def book_store(num: str) -> None:
    match num:
        case '1':
            print("Homodeus")
        case '2':
            print("Damian")
        case '3':
            print("The vegetarian")
        case _:
            print("없는 책입니다")
    print("이용해주셔서 감사합니다")

book_store(input("1, 2, 3, 중 입력"))
```

```python
## guard 조건 ㅡ 판독기

def separation_of_powers(condition:str)-> str:
    match condition:
        case 'make the law':
            return "Legistalive"
        case 'practice the law':
            return "Judicial"
        case 'manage the society':
            return "Executive"
        case _:
            return "Unknown"

result = separation_of_powers(input("function: "))
print("separation of power: ", result )

```
