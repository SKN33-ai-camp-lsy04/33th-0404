## 1. dict 생성과 key 규칙

`dict`의 핵심은 `key`이다.

- 하나의 `dict` 안에서 `key`는 중복될 수 없다.
- 같은 `key`를 다시 넣으면 마지막 값으로 덮어써진다.
- `key`는 값을 빠르게 찾기 위한 기준이므로 중간에 값이 바뀌면 안 된다.
- 그래서 `int`, `float`, `str`, `tuple` 같은 불변 자료형은 `key`로 사용할 수 있지만, `list`, `dict`, `set` 같은 가변 자료형은 `key`로 사용할 수 없다.
- Python 3.7 이상에서 `dict`는 삽입 순서를 유지한다.

```python
print("--- dict 조회 ---")
print("dct: ", dct)

# key를 이용한 조회
print("dct['a']: ", dct['a'])
print("dct.get['a']: ", dct.get('a'))

# 없는 key 조회
# print("dct['d']: ", dct['d'])
# KeyError 발생

print("dct.get['d']: ", dct.get('d'))

# 없으면 기본값 적용
print("dct.get('c', '값없음'): ", dct.get('c', '값없음'))
```

## dict 요소 추가, 수정, 삭제
- dict는 mutable 자료형
- dict[없는 key] = value
- dict[key] = value로 value 수정 가능
- dict.update()를 이용해 value 수정 가능 (여러 값 동시 수정도 가능)
- pop() 또는 del을 이용해 요소 제거

```python
dct['e'] = 888
print(dct)

dct['e'] = 999  # 수정
print(dct)

# update()
dct.update( e = 1234) # 알아서 자료형 판독한다
print(dct)

# 여러 value update
dct.update({'a':111, 'b': 222})

# 요소 제거
# 1) del 연산
del dct['e']
# 2) pop() 메서드 == 꺼내기
temp = dct.pop('b')
print(temp)
print(dct)
```

```python
# dict 생성 함수와 주요 메서드
# - dict 생성 방법
# 1) {k:v, k:v, ...}
# 2) dict(key=value, key=value, ...)

dct3 = dict(name='청가람', age=17)
print(dct3)

# 주요 메서드
keys = dct3.keys()
print(keys)
print(keys, type(keys)) #dict_keys 타입

values = dct3.values()
print(values, type(values))

items = dct3.items() # item(요소 하나 == k:v) 추출
print(items, type (items))
```

```python
import copy # copy 모듈 가져오기

print ("--- copy.deepcopy() : 객체 복제 ---")
fruit = {
    "name" : "apple",
    "price" : "2000"
    "options" : ["red", "200g"]
}

deep_copy_fruit = copy.deepcopy(fruit) # 깊은 복사(객체 복제)
print("id(fruit): ", id(fruit))
print("id(deep_copy_fruit): ", id(deep_copy_fruit))

fruit['price'] = 3000
fruit['options'].append("상주")

print("fruit: ", fruit)
print("deep_copy_fruit: ", deep_copy_fruit)
```
