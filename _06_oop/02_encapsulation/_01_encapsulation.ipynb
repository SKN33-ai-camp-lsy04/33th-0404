## Python private과 name mangling

Python에서 `__balance`처럼 언더스코어 두 개로 시작하는 속성은 외부에서 그대로 접근하기 어렵게 이름이 바뀐다.
이를 name mangling이라고 한다.

다만 이것은 보안을 위한 완전한 접근 제한이라기보다, 클래스 외부에서 직접 사용하지 말라는 강한 신호에 가깝다.
